# KNN_Algorithm

```python
def KNN(d_train,y_train,xq,k):
    distences=[]
    for i in range(len(d_train)):
        xi=d_train[i]
        yi=y_train[i]
        di=np.sqrt(np.sum((xi-xq)**2))
        distences.append([yi,di])
    distences.sort(key=lambda x:x[1])
    k_nn=distences[:k]
    new={}
    for i in k_nn:
        if i[0] in new:
            new[i[0]]=new[i[0]]+1
        else:
            new[i[0]]=1
    l=[]
    for i in new:
        l.append(new[i])
    high=max(l)
    new1=[]
    for k,v in new.items():
        if high==v:
            new1.append(k)
    return new1[0]
d_train= np.array([
    [150, 50],
    [160, 60],
    [170, 65],
    [180, 75]
])
y_train = np.array(['A', 'A', 'B', 'B'])
xq = np.array([165, 62])
k = 3
KNN(d_train,y_train,xq,k)
```

```python
#Accuracy
x_test = np.array([
    [155, 55],
    [165, 60],
    [175, 70]
])
y_test = np.array(['A', 'A', 'B'])
correct=0
for i in range(len(x_test)):
    predicted=KNN(d_train,y_train,x_test[i],k)
    if predicted==y_test[i]:
        correct+=1
accuracy=correct/len(x_test)
accuracy
```

# KD_Tree

```python
def kd_tree(lst,count):
    if len(lst)==0:
        return None
    if count%2==0:
        lst.sort(key=lambda x:x[0])
    else:
        lst.sort(key=lambda x:x[1])
    median=len(lst)//2
    return {
        'Root_node':lst[median],
        'Left_Node':kd_tree(lst[:median],count+1),
        'Right_Node':kd_tree(lst[median+1:],count+1)
    }
#lst=[
 #   (7,2),(5,4),(9,6),(2,3),(4,7),(8,1)
#]
lst=[
    (7,2,3),(5,4,4),(9,6,12),(2,3,11),(4,7,9),(8,1,14)
]
count=0
kd_tree(lst,count)
```

# LSH

```python
def LSH(lst,H1,H2,H3,H4,xq):
    l=[[0,0,0,0] for i in range(len(lst))]
    for i in range(len(lst)):
        p1=((lst[i][0]*H1[0])+(lst[i][1]*H1[1]))
        p2=((lst[i][0]*H2[0])+(lst[i][1]*H2[1]))
        p3=((lst[i][0]*H3[0])+(lst[i][1]*H3[1]))
        p4=((lst[i][0]*H4[0])+(lst[i][1]*H4[1]))
        if p1>0:
            l[i][0]=1
        if p2>0:
            l[i][1]=1
        if p3>0:
            l[i][2]=1
        if p4>0:
            l[i][3]=1
    new={}
    for i in range(len(l)):
        key=tuple(l[i])
        if key not in new:
            new[key]=[]
        new[key].append(i)
    query=[]
    a=((xq[0]*H1[0])+(xq[1]*H1[1]))
    b=((xq[0]*H2[0])+(xq[1]*H2[1]))
    c=((xq[0]*H3[0])+(xq[1]*H3[1]))
    d=((xq[0]*H4[0])+(xq[1]*H4[1]))
    if a>0:
        query.append(1)
    else:
        query.append(0)
    if b>0:
        query.append(1)
    else:
        query.append(0)
    if c>0:
        query.append(1)
    else:
        query.append(0)
    if d>0:
        query.append(1)
    else:
        query.append(0)
    new1=None
    for k,v in new.items():
        if tuple(query)==k:
            new1=v
    xq_mag=np.sqrt((xq[0]**2)+(xq[1]**2))
    NN=[]
    for i in range(len(new1)):
        p_mag=np.sqrt((lst[new1[i]][0]**2)+(lst[new1[i]][1]**2))
        dot_pro=(xq[0]*lst[new1[i]][0])+(xq[1]*lst[new1[i]][1])
        cos_theta=dot_pro/(p_mag*xq_mag)
        NN.append([lst[new1[i]],cos_theta])
    NN.sort(key=lambda x:x[1])
    return 'NearestNeighbour point with cosign_similarity',NN[-1]
lst=[
    [2,3],[3,4],[4,5],[-2,3],[-3,-4],[5,1]
]
H1=[1,1]
H2=[1,-1]
H3=[-1,2]
H4=[-1,-1]
xq=[3,3]
LSH(lst,H1,H2,H3,H4,xq)
```
